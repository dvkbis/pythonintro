1) Créez une fonction nommée calcul_moyenne() qui prend une liste de notes en entrée et retourne la moyenne de ces notes.

```python
def calcul_moyenne(notes):
    moy = 0
    for note in notes:
        moy += note

    return moy / len(notes)

notes_list = [10,12,15,20]
print(f"moy : {calcul_moyenne(notes_list)}")
```

2) Créez une fonction appelée recherche_min() qui prend une liste de nombres en entrée et retourne le plus petit de ces nombres.

```python
def recherche_min(notes):
    if len(notes) == 0: 
        return -1
    
    min = notes[0]
    for note in notes:
        if(note < min):
            min = note

    return min

notes_list = [10,12,5,20]
print(f"min : {recherche_min(notes_list)}")
```

représentant une phrase en entrée et retourne le nombre de mots dans cette
phrase
