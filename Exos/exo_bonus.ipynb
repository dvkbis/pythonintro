```python
from random import randint
import re

def compte_words(sentence: str):
    split_sentence = re.split(r"\s+", sentence)
    words = filter(lambda w: len(w) >= 3, split_sentence)
    words_set = set(list(words))
    return len(words_set)

print(
    "Nombre de mots:",
    compte_words("Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat. Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum.")
)
```

2. Compter les îles (version simple)
## Setup :
- Générer une "carte" (= une grille) avec de l'eau (=0), de la terre (=1)
## Comptez le nombre d'îles :
- = 1 terre avec que de l'eau comme voisins proches (= horizontalement & verticalement, diagonale ignorée

```python
from random import randint

def generer_carte(height, width):
    carte = []
    for i in range(height):
        line = []
        for j in range(width):
            line.append(randint(0,1))
            
        carte.append(line)
    return carte

def afficher_carte(carte):
    for row in carte:
        print('  '.join(map(str, row)))

def compter_iles(carte):
    def verif_ile(x, y):
        if(x < 0 or x >= len(carte)): 
            return 0
        if(y < 0 or y >= len(carte[x])):
            return 0
        if carte[x][y] != 1:
            return 0
        
        carte[x][y] = 2
        verif_ile(x+1, y)
        verif_ile(x-1, y)
        verif_ile(x, y+1)
        verif_ile(x, y-1)

        return 1   

    counter = 0

    for x in range(len(carte)):
        for y in range(len(carte[x])):
            counter += verif_ile(x, y)

    return counter


carte = generer_carte(5,5)  
afficher_carte(carte)
print("Nombre iles:", compter_iles(carte))
```

# GROUP BY COUNT, SUM, AVG
1. setup
- remplir un tableau avec 10 lignes et une structure "dictionnaires dans une liste"
[{"A":"a", "B":1}, {"A":"b", "B":2}, ...]
- colonnes :
    - "A" à remplir aléatoirement avec comme lettres possibles "a", "b", "c"
    - "B" à remplir aléatoirement avec des entiers
2. compter les occurrences
- pour la colonne "A" : on veut compter le nombre de fois qu'un élément est présent dans cette colonne, et ce, pour chaque élément distinct
3. compter la somme et la moyenne des valeurs dans "B" pour la lettre "a" de la colonne "A"

4. compter la somme et la moyenne des valeurs dans "B" pour chaque lettre de la colonne "A"

5. faire la même chose pour la structure 'listes dans dictionnaire':

{"A":["a","b",...], "B":[1,2,...]}

```python
from random import randint

def remplir_dict():
    dict_list = []
    for _ in range(10):
        random_letter = ['a','b','c'][randint(0,2)]
        random_number = randint(1,20)
        dict_list.append({"A": random_letter, "B": random_number})

    return dict_list

def restructure_dict(dict_list: list):
    new_struct = {"A": [], "B": []}
    for my_dict in dict_list:
        new_struct["A"].append( my_dict["A"] )
        new_struct["B"].append( my_dict["B"] )

    return new_struct

my_dict_list = remplir_dict()
my_restructure_list = restructure_dict(my_dict_list)
print(my_dict_list)

# 2. compter les occurrences
def compter_occurences(dict_list: list):
    occurences_dict = {'a': 0, 'b': 0, 'c': 0}
    for my_dict in dict_list:
        letter = my_dict["A"]
        occurences_dict[letter] += 1

    return occurences_dict

occurences = compter_occurences(my_dict_list)
print("ocurences", occurences)

# 3. compter la somme et la moyenne des valeurs dans "B" pour la lettre "a" de la colonne "A"
def compter_moy_somme(dict_list: list, letter_target):
    somme = 0
    n = 0

    for my_dict in dict_list:
        letter = my_dict["A"]
        if letter == letter_target:
            somme += my_dict["B"]
            n += 1
        
    return (somme, somme / n) if n != 0 else (0, 0)

somme_moy = compter_moy_somme(my_dict_list, "a")
print(f"somme {somme_moy[0]} -- moyenne {somme_moy[1]}")

# 4. compter la somme et la moyenne des valeurs dans "B" pour chaque lettre de la colonne "A"
def calcule_moy(dict_by_letter):
    for sum_moy in dict_by_letter.values():
        if sum_moy[1] != 0:
            sum_moy[2] = sum_moy[0] / sum_moy[1]

def compter_moy_somme(dict_list: list):
    by_letter = {"a": [0,0,0], "b": [0,0,0], "c": [0,0,0]}     # "a" : [somme, count, moy]

    for my_dict in dict_list:
        letter = my_dict["A"]

        by_letter[letter][0] += my_dict["B"] # somme
        by_letter[letter][1] += 1 # count

    calcule_moy(by_letter)
        
    return by_letter

somme_moy_list = compter_moy_somme(my_dict_list)
print(somme_moy_list)

print("-----")

# 5. faire la même chose pour la structure 'listes dans dictionnaire':
# {"A":["a","b",...], "B":[1,2,...]}
print(my_restructure_list)

# 2. bis
def compter_occurences_bis(dict_list: list):
    occurences_dict = {"a": 0, "b": 0, "c": 0}
    for lower_letter in dict_list["A"]:
            occurences_dict[lower_letter] += 1

    return occurences_dict

print ("occurences:", compter_occurences_bis(my_restructure_list))
# 3. bis
def compter_moy_somme_bis(dict_list, letter_target):
    somme = 0
    n = 0
    for i in range(len(dict_list["A"])):
        lower_letter = dict_list["A"][i]
        if lower_letter == letter_target:
            somme += dict_list["B"][i]
            n += 1

    return (somme, somme / n) if n != 0 else (0, 0)

somme_moy_list = compter_moy_somme_bis(my_restructure_list, "a")
print(f"somme {somme_moy[0]} -- moyenne {somme_moy[1]}")

# 4. bis
def compter_moy_somme_bis(dict_list):
    by_letter = {"a": [0,0,0], "b": [0,0,0], "c": [0,0,0]}     # "a" : [somme, count, moy]

    for i in range(len(dict_list["A"])):
        lower_letter = dict_list["A"][i]
        number = dict_list["B"][i]
        by_letter[lower_letter][0] += number
        by_letter[lower_letter][1] += 1

    calcule_moy(by_letter)
    return by_letter

somme_moy_list = compter_moy_somme_bis(my_restructure_list)
print(somme_moy_list)

    


            
```

# 5. Fibonacci
- rappel de la suite de Fibonacci:
    - 0,1,1,2,3,5,8,13,...
    - F(n+1) = Fn + F(n-1)

 
- Creer une fonction qui calcule la valeur de la suite pour un 'N' donné.
- Faites (au moins) une fonction pour chaque stratégie:
1. fonction récursive (sans stocker les valeurs déjà parcourues)
2. recursive : f(x) est appelé dans f(x)
    - fonction récursive (en stockant les valeurs déjà pourcourues)
3. boucles (for ou while)

```python
def fibonacci (n):
    if n <= 1:
        return 0
    first = 0
    second = 1

    for _ in range(n):
        last = first + second
        first = second
        second = last
    
    return second

print(fibonacci(100))
```

```python
def fibonacci_recursive(n):
    if n <= 1: return 1

    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

print(fibonacci_recursive(10))
```

```python

def fibonacci_recursive_memo(n):
    memo = [0] * n

    def fibonnacci(n):
        if n <= 1: 
            return 1
        if memo[n] != 0: 
            return memo[n]

        memo[n] = fibonnacci(n - 1) + fibonnacci(n - 2)
        return memo[n]

    return fibonacci(n)
    

print(fibonacci_recursive_memo(100))
```
