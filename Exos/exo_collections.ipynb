3) Générez deux ensembles de nombres aléatoires compris entre 1 et 20. Affichez ces deux ensembles générés. Enfin, trouvez l'intersection des deux ensembles et affichez-la

```python
from random import randint

def generer_list_nombre(min, max, size):
    list = set()
    for i in range(size):
        list.add(randint(min, max))

    return list

list1 = generer_list_nombre(1,20, 4)
list2 = generer_list_nombre(1,20, 5)

print(list1)
print(list2)
print(list1.intersection(list2))
    
```

9) Créez une liste de tuples représentant les commandes d'achat avec les produits et les quantités. Ensuite, créez un dictionnaire de prix pour chaque produit. Calculez ensuite le coût total de toutes les commandes et affichez-le.


```python
from random import randint

produits = ["clavier", "souris", "écran", "ram"]
produits_prix = {}
produits_quantite = []
for produit in produits:
    produits_quantite.append((produit, randint(0,10)))
    produits_prix[produit] = randint(1,25)


produits_cout = []
cout_total = 0


for i in range(len(produits)):
    produit = produits[i]
    quantite = produits_quantite[i][1]
    prix = produits_prix[produit]
    cout = prix * quantite
    cout_total += cout
    print(f"- {produits[i]} ({quantite} x {prix} €) \t| {cout} €")

print("------------------------------")
print(f"TOTAL : \t \t  {cout_total} €")
```

10) Créez une liste de dictionnaires représentant les informations des employés avec leur nom, salaire et département. Calculez la somme des salaires pour chaque département, puis calculez la moyenne des salaires pour chaque département. Enfin, affichez les moyennes des salaires pour chaque département.

```python
# {nom, salaire, dep}
# --> somme des salaires pour chaque dep
# --> moy des salaires par dep
# --> afficher les moy par dep
list_noms = ["Daniel", "Loïc", "Guillaume", "Damien", "David", "Arnaud", "Alexis"]
list_dep =  ["RH", "R&D"]
dict_employes = []
for nom in list_noms:
    dict_employes.append({
        "nom": nom,
        "salaire": randint(1000, 4000),
        "dep": list_dep[ randint(0, len(list_dep) - 1)]
    })

print(dict_employes)

dict_dep = {}

for employe in dict_employes:
    nom, salaire, dep = employe["nom"], employe["salaire"], employe["dep"]

    if dep not in dict_dep:
        dict_dep[dep] = (salaire, 1) 
    else:
        dep_salaire, dep_number = dict_dep[dep] 
        dict_dep[dep] = (dep_salaire + salaire, dep_number + 1)

# Affichage
for dep, values in dict_dep.items():
    print(f"# Département {dep}")
    print(f" - nombre d'employés : {values[1]}")
    print(f" - salaire total : {values[0]} €")
    print(f" - salaire moyen : {round(values[0] / values[1], 2)} €")

```
