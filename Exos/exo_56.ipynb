2. Créez un programme qui convertit une note numérique en une note alphabétique en utilisant une échelle de notation
standard. Utilisez des correspondances pour déterminer la note alphabétique correspondante en fonction de la note
numérique

```python
# A = 90
# B = 80
# C = 70
# D = 60
# F = 0

note_num = 100
note_alpha = "-"
if(note_num <= 100 and note_num >= 90):
    note_alpha = "A"
elif(note_num >= 80):
    note_alpha = "B"
elif(note_num >= 70):
    note_alpha = "C"
elif(note_num >= 60):
    note_alpha = "D"
elif(note_num >= 50):
    note_alpha = "E"
else:
    note_alpha = "F"

print(note_alpha)
```

Créez un programme qui calcule l'indice de masse corporelle (IMC) d'une personne en fonction de son poids et de sa
taille. Utilisez des correspondances pour interpréter et catégoriser l'IMC résultant en différentes catégories de poids.



```python
poids = int(input("Quel est votre poids (en kg): "))
taille = int(input("Quel est votre taille (en cm):")) / 100

imc = poids / (taille * taille)

classification = ""
if imc < 18.5:
    classification = "Insuffisance pondérale (maigreur)"
elif imc < 25:
    classification = "Corpulence normale (poids de forme)"
elif imc < 30:
    classification = "Surpoids"
elif imc < 35:
    classification = "Obésité modérée (classe I)"
elif imc < 40:
    classification = "Obésité sévère (classe II)"
else:
    classification = "Obésité morbide ou massive (classe III)"

print (classification)
```

parmi des options préétablies. Après la sélection, il affiche les choix de l'utilisateur pour chaque repas et résume
l'ensemble des repas de la journée.


```python
menu1 = """# Menu 1 : 
- Petit-déjeuner : Oeufs brouillés
- Déjeuner : Salade composée
- Diner : Soupe
"""
menu2 = """# Menu 2 : 
- Petit-déjeuner : Croissant + café
- Déjeuner : Blanc de poulet
- Diner : Burger
"""
menu3 = """# Menu 3 : 
- Petit-déjeuner : Pancakes 
- Déjeuner : Omelette aux champignons
- Diner : Riz bassmati
"""

print("Choisissez un menu :")
print(menu1)
print(menu2)
print(menu3)
choix = int(input(f"Choix du menu (entre 1 et 3):"))

def valid_choice(menu):
    print("****Menu A Valider**** \n")
    print(menu)
    user_choice = input(f"Valider ce menu (true/false) ?")
    return user_choice == 't' or user_choice.lower() == 'true'

confirmed = False
match choix:
    case 1:
        confirmed = valid_choice(menu1)
    case 2:
        confirmed = valid_choice(menu2)
    case 3:
        confirmed = valid_choice(menu3)
        

if confirmed: 
    print("****Menu Choisi****")
    match choix:
        case 1:
            print(menu1)
        case 2:
            print(menu2)
        case 3:
            print(menu3)
        case _:
            print("**** Aucun menu valable choisi ****")
```
