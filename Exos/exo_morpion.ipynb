## 6. Jeu du morpion (Tic-Tac-Toe, OXO)
### description
Dans une grille de 3 sur 3, deux joueurs s'affrontent pour être le premier joueur à effectuer une ligne de 3 caractères identiques (O pour un joueur, X pour l'autre). La ligne finale peut être horizontale, verticale ou oblique.

Le jeu se déroule en tour par tour.

- Afficher une grille vide.
- Faîtes le placement d’un symbole dans une case souhaitée à l'aide de coordonnées.
- Implémenter la fonction du tour par tour et de restreindre les cases déjà occupées.
- Vérifier les conditions de fin de jeu.
### tips:
- la grille peut se faire avec une liste de liste (ou un dicionnaire)
- implémenter le joueur adverse avec le module random

```python
from IPython.display import clear_output

def afficher_grille(grille):
    clear_output(False)
    for r in grille:
        print(r)
    
def play_user(grille, x, y, player_num):
    if type(x) is not int or type(y) is not int:
        return False
    if x < 0 or x >= 3:
        return False
    if y < 0 or y >= 3:
        return False
    if grille[x][y] != 0:
        return False
    
    grille[x][y] = player_num
    return True

def get_winner_morpion(grille, x, y):
    if not is_in_boundary(x, y):
        return None
    
    checks = [
        count_sequence_horizontal,
        count_sequence_vertical,
        count_sequence_diag_down,
        count_sequence_diag_up
    ]
    for check in checks:
        sequence = check(grille, x, y)
        if len(sequence) == 3:
            return grille[x][y], sequence

    return None


def is_in_boundary(x, y):
    if x < 0 or x >= 3:
        return False
    if y < 0 or y >= 3:
        return False
    
    return True

def count_sequence(grille, x, y, add_x, add_y):
    player_num = grille[x][y]
    x = x + add_x
    y = y + add_y
    same_sequence = set()
    while is_in_boundary(x, y) and grille[x][y] == player_num:
        same_sequence.add((x, y))
        x = x + add_x
        y = y + add_y

    return same_sequence

def count_sequence_vertical(grille, x, y):
    return count_sequence(grille, x, y, 1, 0).union( count_sequence(grille, x, y, -1, 0) ).union({(x,y)})

def count_sequence_horizontal(grille, x, y):
    return count_sequence(grille, x, y, 0, -1).union( count_sequence(grille, x, y, 0, 1) ).union({(x,y)})
    
def count_sequence_diag_down(grille, x, y):
    return count_sequence(grille, x, y, 1, 1).union( count_sequence(grille, x, y, -1, -1) ).union({(x,y)})

def count_sequence_diag_up(grille, x, y):
    return count_sequence(grille, x, y, -1, 1).union( count_sequence(grille, x, y, 1, -1) ).union({(x,y)})
```

### Digit to Position
**FROM numpad**
- 7 8 9
- 4 5 6
- 1 2 3

**TO Points**
- (0, 0) (0, 1) (0, 2)
- (1, 0) (1, 1) (1, 2)
- (2, 0) (2, 1) (2, 2)

 **Convertisseur**
- (0, 0) = 1 --> ((1-1) // 3 = 0, (1-1) % 3 = 0)
- (0, 1) = 2 --> ((2-1) // 2 = 0, (2-1) % 3 = 1)
- (0, 2) = 3 --> ((3-1) // 3 = 0, (3-1) % 3 = 2)
- (1, 0) = 4 --> ((4-1) // 3 = 1, (4-1) % 3 = 1)
- (1, 1) = 5 --> ((5-1) // 3 = 1, (5-1) % 3 = 1)
- (1, 2) = 6 --> ((6-1) // 3 = 1, (6-1) % 3 = 2)
- ...

 **swap**
 - (9, 3)
 - (8, 2)
 - (7, 1)
 - (3, 9)
 - (2, 8)
 - (1, 7)


```python
def swap_digit_from_numpad(digit):
    dict_swap_digit = {1 : 7, 2 : 8, 3 : 9,
                    4 : 4, 5 : 5, 6 : 6,
                    7 : 1, 8 : 2, 9 : 3}
    
    if (type(digit) is int and digit > 0 and digit < 10):
        return dict_swap_digit[digit]
    
    return -1

def convert_digit_to_point(digit):
    if type(digit) is not int:
       return (-1, -1)
    if digit < 1 or digit > 9:
       return (-1, -1)
    
    return ((digit - 1) // 3, (digit - 1) % 3)

def convert_numpad_to_position(numpad_position):
    swapped_digit = swap_digit_from_numpad(numpad_position)
    return convert_digit_to_point(swapped_digit)
```

## Morpion : Player vs Player

```python
def get_player_input(player_number):
    print(f">> Player  {player_number} en attente..")
    # x_player = int(input("x pos (entre 0 et 2)"))
    # y_player = int(input("y pos (entre 0 et 2)"))
    numpad_position = int(input("Choisissez une position à partir du numpad (numpad = position morpion)"))
    return convert_numpad_to_position(numpad_position)

def play_morpion(grille_morpion):
    winner = None
    round = 0
    afficher_grille(grille_morpion)

    while (winner is None and round < 9):
        player_number = (round % 2) + 1
        is_playing = True
        x = -1
        y = -1
        
        while (is_playing):
            x, y = get_player_input(player_number)
            is_playing = not play_user(grille_morpion, x, y, player_number) 
            afficher_grille(grille_morpion)
            if is_playing: 
                print(">>>> Mauvaise entrée !") 

        if round > 3:
            winner = get_winner_morpion(grille_morpion, x, y)
        round += 1
        
    return winner


grille_morpion = [[0,0,0], [0,0,0], [0,0,0]]
winner = play_morpion(grille_morpion)
print (f"Gagnant : {winner}")
```

## Morpion : Player vs Bot (Random)

```python
from IPython.display import clear_output
from random import randint


def find_random_play_bot(grille):
    already_played_set = set()
    x = randint(0, 2)
    y = randint(0, 2)
    
    while len(already_played_set) < 9 and grille[x][y] != 0:
        already_played_set.add((x,y))
        x = randint(0, 2)
        y = randint(0, 2)
    
    return (x, y) if len(already_played_set) < 9 else (-1, -1)

def play_morpion_vs_bot(grille_morpion, bot_play_fn):
    winner = None
    round = 0
    afficher_grille(grille_morpion)

    while (winner is None and round < 9):
        player_id = (round % 2) + 1
        x = -1
        y = -1

        if player_id == 1:
            x, y = get_player_input(player_id)
        else:
            x, y = bot_play_fn(grille_morpion)

        is_correct_play = play_user(grille_morpion, x, y, player_id) 
        afficher_grille(grille_morpion)

        if is_correct_play:
            winner = get_winner_morpion(grille_morpion, x, y)
            round += 1
        else:
            if player_id == 1:
                print(">>>> Mauvaise entrée !")
            else:
                print(">>>> Une erreur est survenue ")
                break

    return winner

grille_morpion = [[0,0,0], [0,0,0], [0,0,0]]
winner = play_morpion_vs_bot(grille_morpion, find_random_play_bot)
print (f"Gagnant : {winner}")
```

## Morpion : Player Vs Bot (Intermediate)

```python
from random import shuffle
from IPython.display import clear_output

def find_winning_pos(grille, num_player):
    grille_size = len(grille)
    for x in range(grille_size):
        for y in range(grille_size):
            winner = None
            if grille[x][y] == 0:
                grille[x][y] = num_player
                winner = get_winner_morpion(grille, x, y)
                grille[x][y] = 0
            if winner is not None:
                return x, y
            
    return -1, -1

def find_play_edge_or_middle(grille):
    possible_play_list = [(0,0), (2,2), (1,1), (0,2), (2,0)]
    shuffle(possible_play_list)

    while len(possible_play_list) > 0:
        x, y = possible_play_list.pop()
        if grille[x][y] == 0:
            return x, y
        
    return -1, -1

def get_intermediate_bot_play(grille_morpion):
    x,y = find_winning_pos(grille_morpion, 2)
    if (x,y) != (-1,-1):
        return x,y    

    x,y = find_winning_pos(grille_morpion, 1)
    if (x,y) != (-1,-1):
        return x,y

    x,y = find_play_edge_or_middle(grille_morpion)

    if (x,y) != (-1, -1):
        return x, y
    
    return find_random_play_bot(grille_morpion)

grille_morpion = [[0,0,0], [0,0,0], [0,0,0]]
winner = play_morpion_vs_bot(grille_morpion, get_intermediate_bot_play)
print (f"Gagnant : {winner}")
    
```
