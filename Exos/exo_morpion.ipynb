## 6. Jeu du morpion (Tic-Tac-Toe, OXO)
### description
Dans une grille de 3 sur 3, deux joueurs s'affrontent pour être le premier joueur à effectuer une ligne de 3 caractères identiques (O pour un joueur, X pour l'autre). La ligne finale peut être horizontale, verticale ou oblique.

Le jeu se déroule en tour par tour.

- Afficher une grille vide.
- Faîtes le placement d’un symbole dans une case souhaitée à l'aide de coordonnées.
- Implémenter la fonction du tour par tour et de restreindre les cases déjà occupées.
- Vérifier les conditions de fin de jeu.
### tips:
- la grille peut se faire avec une liste de liste (ou un dicionnaire)
- implémenter le joueur adverse avec le module random

```python

def afficher_grille(grille):
    clear_output(False)
    for r in grille:
        print(r)
    
def play_user(grille, x, y, player_num):
    if type(x) is not int or type(y) is not int:
        return False
    if x < 0 or x >= 3:
        return False
    if y < 0 or y >= 3:
        return False
    if grille[x][y] != 0:
        return False
    
    grille[x][y] = player_num
    return True

# TODO get_winner_sequence()
def get_winner_morpion(grille, x, y):
    if not is_in_boundary(x, y):
        return set()
    
    # TODO loop fn
    sequence = count_sequence_vertical(grille, x, y)
    if len(sequence) >= 3:
        return sequence

    sequence = count_sequence_horizontal(grille, x, y)
    if len(sequence) >= 3:
        return sequence

    sequence = count_sequence_diag_down(grille, x, y)
    if len(sequence) >= 3:
        return sequence
    
    sequence = count_sequence_diag_up(grille, x, y)
    if len(sequence) >= 3:
        return sequence

    return set()


def is_in_boundary(x, y):
    if x < 0 or x >= 3:
        return False
    if y < 0 or y >= 3:
        return False
    
    return True

def count_sequence(grille, x, y, add_x, add_y):
    player_num = grille[x][y]
    x = x + add_x
    y = y + add_y
    same_sequence = set()
    while is_in_boundary(x, y) and grille[x][y] == player_num:
        same_sequence.add((x, y))
        x = x + add_x
        y = y + add_y

    return same_sequence

def count_sequence_vertical(grille, x, y):
    return count_sequence(grille, x, y, 1, 0).union( count_sequence(grille, x, y, -1, 0) ).union({(x,y)})

def count_sequence_horizontal(grille, x, y):
    return count_sequence(grille, x, y, 0, -1).union( count_sequence(grille, x, y, 0, 1) ).union({(x,y)})
    
def count_sequence_diag_down(grille, x, y):
    return count_sequence(grille, x, y, 1, 1).union( count_sequence(grille, x, y, -1, -1) ).union({(x,y)})

def count_sequence_diag_up(grille, x, y):
    return count_sequence(grille, x, y, -1, 1).union( count_sequence(grille, x, y, 1, -1) ).union({(x,y)})
```

### Digit to Position
**FROM numpad**
- 7 8 9
- 4 5 6
- 1 2 3

**TO Points**
- (0, 0) (0, 1) (0, 2)
- (1, 0) (1, 1) (1, 2)
- (2, 0) (2, 1) (2, 2)

 **Convertisseur**
- (0, 0) = 1 --> ((1-1) // 3 = 0, (1-1) % 3 = 0)
- (0, 1) = 2 --> ((2-1) // 2 = 0, (2-1) % 3 = 1)
- (0, 2) = 3 --> ((3-1) // 3 = 0, (3-1) % 3 = 2)
- (1, 0) = 4 --> ((4-1) // 3 = 1, (4-1) % 3 = 1)
- (1, 1) = 5 --> ((5-1) // 3 = 1, (5-1) % 3 = 1)
- (1, 2) = 6 --> ((6-1) // 3 = 1, (6-1) % 3 = 2)
- ...

 **swap**
 - (9, 3)
 - (8, 2)
 - (7, 1)
 - (3, 9)
 - (2, 8)
 - (1, 7)


```python
def swap_digit_from_numpad(digit):
    dict_swap_digit = {1 : 7, 2 : 8, 3 : 9,
                    4 : 4, 5 : 5, 6 : 6,
                    7 : 1, 8 : 2, 9 : 3}
    
    if (type(digit) is int and digit > 0 and digit < 10):
        return dict_swap_digit[digit]
    
    return -1

def convert_digit_to_point(digit):
    if type(digit) is not int:
       return (-1, -1)
    if digit < 1 or digit > 9:
       return (-1, -1)
    
    return ((digit - 1) // 3, (digit - 1) % 3)

def convert_numpad_to_position(numpad_position):
    swapped_digit = swap_digit_from_numpad(numpad_position)
    return convert_digit_to_point(swapped_digit)
```
