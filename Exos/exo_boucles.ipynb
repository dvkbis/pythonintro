Concevez un programme qui génère et affiche les nombres premiers jusqu'à
100 en utilisant une boucle.

```python
from math import sqrt
prime_list = list()
for n in range(1,101):
    i = 2
    isPrime = True
    while(i < sqrt(n)):
        if n % i == 0:
            isPrime = False
        i+=1
    if isPrime: 
        prime_list.append(n)

print(prime_list)

```
