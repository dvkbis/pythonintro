### Try writing regexes to:

1. Extract all names.
2. Extract all email addresses.
3. Extract only the domains (example.com, gmail.com, etc.).
4. Replace all email addresses with [REDACTED].

```python
import re

text = """
Alice: alice@example.com 
Bob: bob@gmail.com
Charlie: charlie@yahoo.com
"""
## other way ::
r = re.compile(r"\w+@\w+\.\w+")
r.findall(text)
## group name
# a = re.findall(r"(?P<name>\w+)@(?P<domain>\w+\.\w+)", text)
result_name = re.findall(r"(\w+)@\w+\.\w+", text)
result_domain = re.findall(r"\w+@(\w+\.\w+)", text)
result_email = re.findall(r"\w+@\w+\.\w+", text)
result_replace = re.sub(r"\w+@\w+\.\w+", "REDACTED", text)

print(result_name)
print(result_domain)
print(result_email)
print(result_replace)

```

### Replace name and show the first char

```python
import re

text = """
Alice: alice@example.com 
Bob: bob@gmail.com
Charlie: charlie@yahoo.com
"""

result = re.sub(
    r"(\w{1})([\w.-]+)@([\w.-]+\.\w+)",
    r"\1***@\3",
    text
)

print(result)
```

```python
import re
text = """
Alice: alice@example.com 
Bob: bob@gmail.com
Charlie: charlie@yahoo.com
"""
def hide_name(match):
    name = match.group(1)
    domain = match.group(2)

    return name[0] + "*" * len(name[1:]) +  "@" + domain

result = re.sub(
    r"([\w.-]+)@([\w.-]+\.\w+)",
    hide_name,
    text
)

print(result)
```
