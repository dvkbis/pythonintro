- Vous devez créer un programme Python permettant de gérer une
liste de tâches.
- L’objectif est de simuler le fonctionnement d’une application réelle
avec des fonctionnalités complètes : ajout, modification,
suppression, filtrage et pagination.

```python
tasks = [
{
"id": 1,
"title": "Coder le projet",
"status": "todo",
"priority": 3,
"tags": ["dev", "urgent"]
}
]

## Créer une fonction permettant d’ajouter une tâche à la liste.
#  Générer automatiquement un id unique
# * Ajouter la tâche avec le status par défaut "todo"
id = 0
def add_task(tasks, title, priority, tags):
    global id 
    id += 1
    tasks.append({"id": id, "title": title, "status": "todo", "priority": priority, "tags": tags})

## Créer une fonction pour afficher les tâches avec filtres et pagination.
# params est un dictionnaire optionnel pouvant contenir :
# * "status" : filtre par statut
# * "priority_min" : priorité minimum
# * "tag" : filtre par tag
# * "page" : numéro de page (10 éléments par page)
# Exemples :
# list_tasks(tasks, {"status": "todo"})
# list_tasks(tasks, {"priority_min": 3, "page": 2})
# list_tasks(tasks, {"tag": "dev"})
# Affichage attendu :
# id - titre - status - priorité - tags
def list_tasks(tasks, params):
    per_page = 10
    result = tasks
    for key, value in params:
        if key in ("status", "tag"):
            result = filter_tasks(result, key, value)
        elif key == "priority_min":
            result = filter_min_priority(result, value)
    if "page" in params:
        starting_index = (per_page * params["page"]) - 1
        return result[starting_index: starting_index + per_page]
    
    return result


def filter_tasks(tasks, key, value):
    result = []
    for task in tasks:
        if task[key] == value:
            result.append(task)

    return task

def filter_min_priority(tasks,value):
    result = []
    for task in tasks:
        if task["priority"] >= value:
            result.append(task)

    return task    

def print_task(tasks):
    print(f"id |\t titre \t|\t status \t|\t priorité |  tags ")
    print("-------------------------------------------------------------------------")
    for task in tasks:
        print(f"{task["id"]} |\t {task["title"]}\t|\t {task["status"]}    \t|\t {task["priority"]}   \t| {task["tags"]} ")

# Créer une fonction pour modifier une tâche existante.
# update_task(tasks, task_id, new_data)
# new_data est un dictionnaire pouvant contenir :
# * "title"
# * "status"
# * "priority"
# * "tags"
def update_task(tasks, task_id, new_data):
    index = find_index_task(tasks, task_id)
    if index >= 0 :
        for key, new_value in new_data.items():
            tasks[index][key] = new_value
        return True
    return False

## Créer une fonction pour supprimer une tâche.
def delete_task(tasks, task_id):
    index = find_index_task(tasks, task_id)
    if index >= 0 :
        tasks.pop(index)
        return True
    return False

def find_index_task(tasks, task_id):
    size = len(tasks)
    found = False
    i = 0
    while i < size and not found:
        if tasks[i]["id"] == task_id:
            return i
        i += 1

    return -1

def mark_done(tasks, task_id):
    index = find_index_task(tasks, task_id)
    if index >= 0:
        tasks[index]["status"] = "done"
        return True
    return False


def menu_tasks():
    print("1/ Add task")
    print("2/ Delete task")
    print("3/ Update task")
    print("4/ List task")
    print("5/ Filter task")
    print("6/ Write")
    command_user = int(input("Entrez une commande: "))
    match command_user:
        case 1:
            new_task = get_input_add()
            add_task(tasks, new_task["title"], new_task["priority"], new_task["tags"])
        case 2:
            task_id = int(input("Entrez une id: "))    
            delete_task(tasks, task_id)
        case 3:
            task_id = int(input("Entrez une id: "))    
            update_task(tasks, task_id)
        case 4:
            print_task(tasks)
        case 5:
            pass


def get_input_add():
    title = input("Add a Title")
    priority = int(input("Add a priority"))
    tags = input("Add tags (dev, urgent)")
    
    return {"title": title, "priority": priority, "tags": tags}


my_tasks = []
add_task(my_tasks, "Test", 3, ["Dev","Urgent"])
add_task(my_tasks, "Java", 2, ["Dev"])
add_task(my_tasks, "Code", 1, ["Perso"])

delete_task(my_tasks, 2)
update_task(my_tasks, 3, {"priority":2, "title": "Odoo"})
mark_done(my_tasks, 3)
print_task(my_tasks)

```
