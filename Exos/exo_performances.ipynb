```python
import timeit

def inverser_chaine(chaine):
    return chaine[::-1]

def inverser_chaine_loop(chaine):
    inversed = ""
    for i in range(len(chaine)-1, -1, -1):
        inversed += chaine[i]
    return inversed

print(inverser_chaine("Hello World!"))
temps_inverser_basic = timeit.timeit(
    stmt= lambda: inverser_chaine("Hello World!"),
    number=1000
)
temps_inverser_loop = timeit.timeit(
    stmt= lambda: inverser_chaine_loop("Hello World!"),
    number=1000
)
print("temps total (normal):", temps_inverser_basic)
print("temps total (loop):", temps_inverser_loop)

```

```python
import tracemalloc
def inverser_chaine(chaine):
    return chaine[::-1]

def inverser_chaine_loop(chaine):
    inversed = ""
    for i in range(len(chaine)-1, -1, -1):
        inversed += chaine[i]
    return inversed

tracemalloc.start()
for _ in range(1000):
    inverser_chaine("Hello World! This is a test!")
_, pic_inverser = tracemalloc.get_traced_memory()
tracemalloc.stop()

tracemalloc.start()
for _ in range(1000): 
    inverser_chaine_loop("Hello World! This is a test!")
_, pic_inverser_loop = tracemalloc.get_traced_memory()
tracemalloc.stop()

print(f"{pic_inverser / 1_000_000:.4f} MB")
print(f"{pic_inverser_loop / 1_000_000:.4f} MB")
```
