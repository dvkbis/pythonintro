
 Ecris une fonction group_anagrams(words) qui prend une liste de mots et retourne une liste de
groupes d'anagrammes. Deux mots sont des anagrammes s'ils contiennent les memes lettres
dans un ordre different.
```
    group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"])
    # [["eat", "tea", "ate"], ["tan", "nat"], ["bat"]]
    group_anagrams(["abc", "bca", "xyz", "zyx", "cab"])
    # [["abc", "bca", "cab"], ["xyz", "zyx"]]
```
L'ordre des groupes et l'ordre a l'interieur des groupes n'ont pas d'importance.
Cas limites a gerer : liste vide, un seul mot, mots de longueurs differentes, mots en double.


````python
def group_anagrams(words):
    dict_anagram = {}
    for word in words:
        word_visited = False
        for key, value_set in dict_anagram.items():
            if is_anagram(word, key):
                value_set.add(word)
                word_visited = True
                continue
        if not word_visited:
            dict_anagram[word] = {word}
        
    return dict_anagram
def add_count_char(char, dict_char_count):
    count = 1
    if dict_char_count.get(char):
        count += dict_char_count[char]
    
    dict_char_count[char] = count


def is_anagram(first, second):
    if len(first) != len(second):
        return False
    
    if first == second:
        return True
    
    first_char_count = {}
    second_char_count = {}
    for i in range(len(first)):
        add_count_char(first[i], first_char_count)
        add_count_char(second[i], second_char_count)
    
    return first_char_count == second_char_count

print(group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"]))
print(group_anagrams(["abc", "bca", "xyz", "zyx", "cab"]))
    # [["abc", "bca", "cab"], ["xyz", "zyx"]]
````
