 Ecris une fonction compress(s) qui prend une chaine de caracteres et retourne sa version
compressee. Chaque groupe de caracteres identiques consecutifs est remplace par le caractere
suivi du nombre de repetitions. Si un caractere apparait une seule fois, on ne met pas de chiffre.

`´ `
 compress("aaabbc") # "a3b2c"
 compress("abcd") # "abcd"
 compress("aaa") # "a3"
 compress("aabbbcccc") # "a2b3c4"
`´`

 Cas limites a gerer : chaine vide, un seul caractere, tous les caracteres identiques, aucune
repetition

```python
import tracemalloc

def compress(s):
    if len(s) == 0 or len(s) == 1:
        return s
    length = len(s)
    result = s[0]
    count = 1
    i = 1

    while i < length:
        current = s[i]
        if result[-1] == current:
            count += 1
        elif count > 1:
            result += f"{count}{current}"
            count = 1
        else:
            result += current
        i+=1

    if count > 1:
        result += f"{count}"

    return result     


def compress_b(s):
    if len(s) == 0 or len(s) == 1:
        return s
    length = len(s)
    result = [s[0]]
    count = 1
    i = 1

    while i < length:
        current = s[i]
        if result[-1] == current:
            count += 1
        elif count > 1:
            result.append(f"{count}")
            result.append(current)
            count = 1
        else:
            result.append(current)
        i+=1

    if count > 1:
        result.append(f"{count}")

    return "".join(result) 
# print("a3b2c -> aaabbc", compress("aaabbc"))
# print("abcd -> abcd ?", compress("abcd")) # "abcd"
# print("aaa -> a3 ?", compress("aaa")) # "a3"
# print("aabbbcccc -> a2b3c4 ?", compress("aabbbcccc")) # "a2b3c4"
# print("aabbbcccc -> a2b3c4 ?", compress("aaa111")) # "a2b3c4"



tracemalloc.start()
compress("a" * 1000000)
mem_a = tracemalloc.get_tracemalloc_memory()
tracemalloc.stop()

tracemalloc.start()
compress_b("a" * 1000000)
mem_b =  tracemalloc.get_tracemalloc_memory()
tracemalloc.stop()

print(mem_a)
print(mem_b)
```
