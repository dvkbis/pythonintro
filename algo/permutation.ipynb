```python
class Solution:
    def permute(self, s: str):
        res = []
        path = []
        used = set()

        def backtrack():
            if len(path) == len(s):
                res.append("".join(path))
                return

            for i in range(len(s)):
                if i in used:
                    continue

                used.add(i)
                path.append(s[i])

                backtrack()

                path.pop()
                used.remove(i)

        backtrack()
        return res
```

```python
"""
liste booléenne
"""
class Solution:
    def permute(self, s: str):
        res = []
        path = []
        used = [False] * len(s)

        def backtrack():
            if len(path) == len(s):
                res.append("".join(path))
                return

            for i in range(len(s)):
                if used[i]:
                    continue

                used[i] = True
                path.append(s[i])

                backtrack()

                path.pop()
                used[i] = False

        backtrack()
        return res
```

```python
""" 
## Variante encore plus “propre” (swap in-place)

Celle-ci est très appréciée en entretien car O(1) mémoire auxiliaire (hors output) :
"""
class Solution:
    def permute(self, s: str):
        arr = list(s)
        res = []

        def backtrack(i):
            if i == len(arr):
                res.append("".join(arr))
                return

            for j in range(i, len(arr)):
                arr[i], arr[j] = arr[j], arr[i]
                backtrack(i + 1)
                arr[i], arr[j] = arr[j], arr[i]

        backtrack(0)
        return res
```
