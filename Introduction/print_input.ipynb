# La fonction **print** et **input**
- **print()** : permet d'affiher du contenu
- **input()** : qui permet d'affiher et de récup une info

```python
print("hello python")
input_user = input("Nom : ")
print(input_user)
```

# Print()
- sep : ajoute un séparateur
- end : ajute un element à la fin

```python
print('hello', 'world', 'python', sep='-',end='!')
```
