# Méthodes du type str

```python
mon_text = "   hello World  "

# Position de la première occurence, Renvoie -1 si non trouvé
print(mon_text.find("o"))

# Renvoie la première occurence enpartant de la fin
print(mon_text.rfind("o"))

# Position de la première occurence, Renvoie une erreur si non trouvé
print(mon_text.index("o"))

# Position de la première occurence en partant de la fin, Renvoie une erreur si non trouvé
print(mon_text.rindex("o"))

# Nombre d'occurences
print(mon_text.count("l"))

# startwith() and endwith()
print(mon_text.startswith("hello"))
print(mon_text.endswith("world"))

# in
if "lo" in mon_text:
    print("in")
else:
    print("not in")

# uppercase() / lowercase() / capitalize
print(mon_text.upper())
print(mon_text.lower())
print(mon_text.capitalize())

# title() : premiere de chaque mot en majuscule
print(mon_text.title())

# swapcase() : inverser maj / min
print(mon_text.swapcase())

# strip() : supprimer des espaces avant et après le string
print(mon_text.strip())
print(mon_text.lstrip())
print(mon_text.rstrip())

# replace
print(mon_text.replace("o", "0"))

# removeprefix()
print(mon_text.removeprefix("   hello"))

```

# Découpage

```python
# split() 
split_text = "un deux trois quatre"
print(split_text.strip().split(" ", 2))
print(split_text.strip().rsplit(" ", 2))

# splitlines()
split_text = "un\ndeux\ntrois\nquatre"
print(split_text.splitlines())

# join
" , ".join(['a','b','c'])

# Découpe en 3 élèments 
email = "admin@email.com"
print(email.partition("@"))
print(email.rpartition("i"))

path = r"c:\\User\Windows\icon"
print(path.rpartition("\\"))
```

# Test

```python
# isAlpha() / # isDigit() / isnumeric() / isalnum() / isspace()

value_str = "Hello"
value_int = "123"
value_notalphanum = "HelloWorld!"
value_allspace = "    "
value_digit = "2"

print(value_str, "is alpha :", value_str.isalpha())
print(value_int, "is numeric :", value_int.isnumeric())
print(value_notalphanum, "is alpha numeric:", value_notalphanum.isalnum())
print(value_allspace, "is space:", value_allspace.isspace())
print(value_digit, "is space:", value_digit.isdigit())
```
