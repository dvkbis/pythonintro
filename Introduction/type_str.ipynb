# Méthodes du type str

```python
mon_text = "hello World"

# Position de la première occurence, Renvoie -1 si non trouvé
print(mon_text.find("o"))

# Renvoie la première occurence enpartant de la fin
print(mon_text.rfind("o"))

# Position de la première occurence, Renvoie une erreur si non trouvé
print(mon_text.index("o"))

# Position de la première occurence en partant de la fin, Renvoie une erreur si non trouvé
print(mon_text.rindex("o"))

# Nombre d'occurences
print(mon_text.count("l"))

# startwith() and endwith()
print(mon_text.startswith("hello"))
print(mon_text.endswith("world"))

# in
if "lo" in mon_text:
    print("in")
else:
    print("not in")

# uppercase() / lowercase() / capitalize
print(mon_text.upper())
print(mon_text.lower())
print(mon_text.capitalize())

# title() : premiere de chaque mot en majuscule
print(mon_text.title())

# swapcase() : inverser maj / min
print(mon_text.swapcase())
```
