# Boucles

```python
# 1. While Loop - continues until condition is False
print("=== While Loop ===")
counter = 0
while counter < 3:
    print(f"Counter: {counter}")
    counter += 1

# 2. For Loop with range() - iterates over numbers
print("\n=== For Loop with range() ===")
for num in range(5):
    print(f"Number: {num}")

# 3. For Loop with enumerate() - gets index and value
print("\n=== For Loop with enumerate() ===")
fruits = ['apple', 'banana', 'cherry']
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")

# 4. For Loop with in - iterates directly over collection
print("\n=== For Loop with in ===")
for item in collection:
    print(f"Item: {item}")

# 5. Nested Loops - loop inside a loop
print("\n=== Nested Loops ===")
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")

# 6. Loop with break - exit loop early
print("\n=== Loop with break ===")
for num in range(10):
    if num == 5:
        break
    print(num)

# 7. Loop with continue - skip current iteration
print("\n=== Loop with continue ===")
for num in range(5):
    if num == 2:
        continue
    print(num)

# 8. range(start, stop, step)
# stop param value excluded
collection = [1,2,3,4,5,6]
for i in range(len(collection) - 1, -1, -1):
    print(collection[i])
```
