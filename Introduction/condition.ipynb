# Match Structure

````python
def describe_value(value):
    match value:
        case 0:
            return "Zero"
        case 1 | 2:
            return "One or Two"
        case int() as number if number > 2:
            return "A positive integer greater than two"
        case str() as text:
            return f"A string: {text}"
        case _:
            return "Something else"

for test_value in [0, 1, 5, "hello", 3.14]:
    print(test_value, "->", describe_value(test_value))
````

# if-ifel-if structure

Use an `if` / `elif` / `else` chain in Python like this:

```python
def evaluate_number(n):
    if n < 0:
        return "Negative"
    elif n == 0:
        return "Zero"
    elif n < 10:
        return "Small positive"
    else:
        return "Large positive"

for value in [-5, 0, 3, 12]:
    print(value, "->", evaluate_number(value))
```
