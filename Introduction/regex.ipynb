## Common Regex Symbols
    .	Any character except newline	c.t → cat, cot, cut
    \d	Digit (0-9)	\d\d → 42
    \w	Letter, digit, or underscore	\w+
    \s	Whitespace	spaces, tabs
    +	One or more	a+
    *	Zero or more	a*
    ?	Optional	colou?r
    ^	Start of string	^Hello
    $	End of string	world$
    \b  Word boundary

```python
re.search(r"c.t", "cat")   # Match
re.search(r"c.t", "cot")   # Match
re.search(r"c.t", "cut")   # Match
```

## Finding All Matches

```python
import re

text = "I have 3 apples and 12 bananas."

numbers = re.findall(r"\d+", text)

print(numbers)
```

##  Character Sets

```python
re.findall(r"[aeiou]", "hello world")
```

## Groups

```python
import re

text = "John: 25"

match = re.search(r"(\w+): (\d+)", text)

print(match.group(1))
print(match.group(2))
```

##  Replacing Text

```python
import re

text = "My phone number is 123-456-7890"

result = re.sub(r"\d", "*", text)

print(result)
```

## Find All

```python
text = "python pytest pyramid java"

words = re.findall(r"\bpy\w*", text)

print(words)
```

## Capturing groups

```python
import re

email = "alice@example.com"

result = re.sub(
    r"(\w+)@(\w+\.\w+)",
    r"***@\2",
    email
)

print(result)
```

## Capturing groups with function

```python
def hide_name(match):
    name = match.group(1)
    domain = match.group(2)

    return "*" * len(name) + "@" + domain

text = """
Alice: alice@example.com
Bob: bob@gmail.com
Charlie: charlie@yahoo.com
"""


result = re.sub(
    r"([\w.-]+)@([\w.-]+\.\w+)",
    hide_name,
    text
)

print (result)
```

## FindIter()

```python
import re

text = "I have 3 apples and 12 bananas"

matches = re.finditer(r"\d+", text)

for match in matches:
    print(match.group())
```

## Flags (IGNORECASE)

| Flag   | Meaning                                     |
| ------ | ------------------------------------------- |
| `re.I` | Ignore case                                 |
| `re.M` | Multiline (`^` and `$` work per line)       |
| `re.S` | Dot matches newlines                        |
| `re.X` | Verbose regex (allows comments and spacing) |


```python
import re

text = "Python PYTHON python"

matches = re.findall(r"python", text, re.IGNORECASE)

print(matches)

# With compiled patterns
pattern = re.compile(r"python", re.I)

print(pattern.findall("Python PYTHON python"))

# with flag
re.findall(r"(?i)python", text)

# Multiple flag
re.findall(
    r"python",
    text,
    re.IGNORECASE | re.MULTILINE
)
```
