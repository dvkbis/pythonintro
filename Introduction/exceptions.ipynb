```python
try:
    input_user = int(12)
    raise Exception()       # equivalent throw
except ValueError:
    print("Erreur de valeur")
except Exception:
    print("Erreur survenue") # - Terminer avec l'exception la plus générique (ici Exception)
else:
    print("execution du code")
finally: 
    print("finally")
```

### Messages d'erreur

```python
try:
    input_user = int("ab")
except ValueError as e:
    print(f"Une erreur de valeur est survenue - {e}")
```

### Plusieurs exception dans un except

```python
try:
    input_user = int("ab")
except (ValueError, TypeError, IndexError):
    print("Une erreur est survenue")
```

### Déclencher une erreur volontairement avec **raise**

```python
try:
    raise Exception("Mon erreur")
except Exception as e:
    print(f"Une erreur est survenue - {e}")
```

### exception DivideByZero

```python
try:
    input_user_1 = int("2")
    input_user_2 = int("0")
    result = input_user_1 / input_user_2
except ZeroDivisionError as e:
    print(f"Une erreur est survenue - {e}")
```

### Créer sa propre Exception

```python
class MyException(Exception): # Héritage
    pass

class MySecondException(Exception):
    def __init__(self, message, action, name):
        self.message = message
        self.action = action
        self.name = name

try:
    input_user = int("1")
    if input_user == 1:
        raise MyException("Une erreur est survenue (Custom Exception)")
    elif input_user == 2:
        raise MySecondException("Seconde exception déclenchée", "Get", "admin")
 
except MyException as e:
    print(f"Une erreur est survenue - {e}")
except MySecondException as e:
    print(f"Une erreur est survenue - {e.message} -- {e.action} : {e.name}")
```

### With Gestion des ressoures
 - Ressource automatique fermée

```python
file = open("text.txt", "w")

file.write("Hello")

file.close()
```

```python
read_file = open("text.txt", "r")
for ligne in read_file.read():
    print(ligne)

read_file.close()
```

```python
with open("text.txt", "w") as f:
    f.write("hello \n")
    f.write("world \n")

with open("text.txt") as f:
    content = f.read()
    print(content)
```
