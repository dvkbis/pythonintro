# Opérateur logiques

- or

- ^ : Ou exclusif 

## Precedence Order
^ is evaluated first (bitwise XOR).
and is evaluated next.
or is evaluated last.

```python
age = 35
nom = "Doe"
role = "Admin"
is_correct = age > 18 and len(nom) and role == "Admin"
print(is_correct)
```

```python
print(True ^ False or True ^ True)
```

```python
# if - elif - else
x = 2
if x > 10:
    print("greater")
elif x == 10:
    print("equal")
else:
    print("lower")
```
