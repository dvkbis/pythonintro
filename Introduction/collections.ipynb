# Les Collections 

##  Types de collections


```python

# liste
list_example = [10, 20, 30]
print("Liste:", list_example)
new_copy_list = list(list_example)
list_example.append(40)
print("Après ajout:", list_example)

# tuple
tuple_example = ("pomme", "banane", "cerise")
print("Tuple:", tuple_example)
print("Premier élément du tuple:", tuple_example[0])

# set
set_example = {"rouge", "vert", "bleu"}
print("Set:", set_example)
set_example.add("jaune")
print("Après ajout:", set_example)

# dictionnaire
dict_example = {"nom": "Alice", "âge": 28, "ville": "Paris"}
print("Dictionnaire:", dict_example)
print("Valeur pour 'nom':", dict_example["nom"])
```

## Methodes list

```python
list_example = [10, 20, 30]

# Examples of common list methods (uses list_example defined earlier)
orig = list_example.copy()
print("Original:", orig)

# append
a = orig.copy()
a.append(50)
print("append(50):", a)

# extend
b = orig.copy()
b.extend([60, 70])
print("extend([60, 70]):", b)

# insert
c = orig.copy()
c.insert(1, 15)
print("insert(1, 15):", c)

# remove
d = orig.copy()
d.remove(20)  # removes first occurrence of 20
print("remove(20):", d)

# pop
e = orig.copy()
last = e.pop()
print("pop() ->", last, "; list:", e)
e2 = orig.copy()
idx_val = e2.pop(1) # removes element at index and return 
print("pop(1) ->", idx_val, "; list:", e2)

# clear
f = orig.copy()
f.clear()
print("clear() ->", f)

# index
print("index(30):", orig.index(30))

# count
print("count(20):", orig.count(20))

# sort
g = [3, 1, 2]
g.sort()
print("sort() on [3,1,2]:", g)

# reverse
h = orig.copy()
h.reverse()
print("reverse():", h)

# copy
i = orig.copy()
print("copy():", i)
```

# Tuples 
- indexée
- immutable
- ordonnée
- hétérogène
- accepte les doublons
- plus rapide que la liste

```python
mon_tuple = 1,2,3

# Tuple unpacking
x, y, z = mon_tuple
print(f"Unpacked: x={x}, y={y}, z={z}")

# Tuple concatenation
tuple1 = (1, 2)
tuple2 = (3, 4)
combined = tuple1 + tuple2
print(f"Concatenation: {combined}")

# Tuple repetition
repeated = (1, 2) * 3
print(f"Repetition: {repeated}")

# Tuple in a loop
for item in mon_tuple:
    print(f"Item: {item}")

# Tuple methods
print(f"Count of 2: {mon_tuple.count(2)}")
print(f"Index of 3: {mon_tuple.index(3)}")

# Nested tuple
nested = (1, (2, 3), 4)
print(f"Nested tuple: {nested}")
print(f"Access nested: {nested[1][0]}")
```
