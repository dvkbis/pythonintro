## Déclaration & affectation
- signe = pour affecter



```python
# Le typage est dynamique
ma_var = 5
ma_var = 'hello'
ma_var = list()
```

```python
# Affectation multiple
a = b = c = 10
print(a)
print(b)
print(c)

# inverser des variables
d = 10
e = 20
d, e = e, d
print(d)
print(e)
```
