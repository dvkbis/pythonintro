```python
import os
import pg8000

# New cell (index 0)
# Example using pg8000 (pure-Python PostgreSQL driver)
# pip install pg8000


conn = pg8000.connect(
    user=os.getenv("PGUSER", "postgres"),
    password=os.getenv("PGPASSWORD", "password"),
    host=os.getenv("PGHOST", "localhost"),
    port=int(os.getenv("PGPORT", 5432)),
    database=os.getenv("PGDATABASE", "testdb"),
)

try:
    cur = conn.cursor()
    cur.execute("SELECT version();")
    print("Postgres version:", cur.fetchone()[0])

    cur.execute("""
    CREATE TABLE IF NOT EXISTS demo (
        id SERIAL PRIMARY KEY,
        name TEXT NOT NULL,
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    );
    """)
    cur.execute("INSERT INTO demo (name) VALUES (%s) RETURNING id;", ("Alice",))
    new_id = cur.fetchone()[0]
    print("Inserted id:", new_id)

    cur.execute("SELECT id, name, created_at FROM demo ORDER BY id DESC LIMIT 5;")
    rows = cur.fetchall()
    for r in rows:
        print(r)

    conn.commit()
finally:
    conn.close()

# https://alembic.sqlalchemy.org/en/latest/ (migration with sql alchemy)
```
