```python
class Solution:
    def repeatedSubstringPattern(self, s: str) -> bool:
        substring = ""

        for i in range(len(s) // 2):
            substring += s[i] 

            if (len(s) % len(substring)) != 0:
                continue

            start = i + 1 
            end = (i + 1) + i + 1 
            is_substring = True
            while end <= len(s) and is_substring:
                if substring != s[start:end]:
                    is_substring = False

                start += i + 1
                end += i + 1
            
            if is_substring:
                return True

        return False
        
```

```python
## Version attendue

class Solution:
    def repeatedSubstringPattern(self, s: str) -> bool:
        n = len(s)

        for pattern_len in range(1, n // 2 + 1):

            if n % pattern_len != 0:
                continue

            pattern = s[:pattern_len]

            if pattern * (n // pattern_len) == s:
                return True

        return False
```
