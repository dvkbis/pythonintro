```python
import re
class Solution:
    def simplifyPath(self, path: str) -> str:
        result = []
        matches = re.findall(r"[^/]+", path)

        for i in range(len(matches)):
            if matches[i] == ".":
                continue
            elif matches[i] == ".." :
                if len(result) > 0:
                    result.pop()
            else:
                result.append(matches[i])

        return "/" + "/".join(result)
```
