```python
def uniquePaths_recursive( m: int, n: int) -> int:
    memo = {}

    def count_path(x, y):
        if x > n or y > m:
            return 0
        if x == n-1 and y == m - 1:
            return 1
        if (x,y) in memo:
            return memo[(x,y)]

        count = 0
        count += count_path(x+1, y)
        count += count_path(x, y+1)

        memo[(x,y)] = count
        return count
    
    return count_path(0, 0)

# print(uniquePaths(4,6))
# 1 1 1
# 1 2 3
# 1 3 6
# 1 4 10

def uniquePaths( m: int, n: int) -> int:
    dp = [ [0]*m for _ in range(n)]

    for i in range(n):
        dp[i][0] = 1

    for i in range(m):
        dp[0][i] = 1


    for i in range(1, n):
        for j in range(1, m):
            dp[i][j] = dp[i-1][j] + dp[i][j-1]
    

    return dp[-1][-1]
            


def uniquePaths_less_memory( m: int, n: int) -> int:
    dp = [1] * m
    
    for _ in range(1, n):
        for i in range(1, m):
            prev = dp[i]
            dp[i] = prev + dp[i-1]

    return dp[-1]

            
print(uniquePaths_less_memory(6,5))
print(uniquePaths(6,5))
print(uniquePaths_recursive(6,5))
```
