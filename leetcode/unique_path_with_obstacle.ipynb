```python
def uniquePathsWithObstacles(obstacleGrid: List[List[int]]) -> int:
    m = len(obstacleGrid)
    n = len(obstacleGrid[0])

    dp = [0] * n
    dp[0] = 1

    for i in range(m):
        if obstacleGrid[i][0] == 1:
            dp[0] = 0
        for j in range(1, n):
            if obstacleGrid[i][j] == 1:
                dp[j] = 0
            else:
                dp[j] = dp[j - 1] + dp[j]

    return dp[-1]
            
```
