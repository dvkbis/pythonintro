```python
class Solution:
    def jump(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return 0

        count = 1
        max_reach = nums[0] 
        next_reach = 0
        i = 1
        while max_reach < (len(nums) - 1):  
            next_reach = max(nums[i] + i, next_reach)
            if i == max_reach:
                max_reach = next_reach
                count += 1
            i += 1

        return count


            

            
        
```

```python
# Solution Attendue : Greedy
class Solution:
    def jump(self, nums: List[int]) -> int:
        jumps = 0
        current_end = 0
        farthest = 0

        for i in range(len(nums) - 1):
            farthest = max(farthest, i + nums[i])

            if i == current_end:
                jumps += 1
                current_end = farthest

        return jumps
```
