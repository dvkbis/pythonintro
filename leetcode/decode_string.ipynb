```python
import re
class Solution:
    def decodeString(self, s: str) -> str:
        pattern = re.compile(r"[a-z]+|\]|\d+")
        matches = pattern.finditer(s)

        result = ""
        counters = []
        words = []

        for match in matches:
            word = match.group()

            if word.isnumeric():
                counters.append(int(word))
                words.append("")
                continue
            if word == ']':
                c = counters.pop()
                w = words.pop()
                word = w * c           

            if len(counters) == 0:
                result += word
            else:
                words[-1] += word


        return result
        
```
