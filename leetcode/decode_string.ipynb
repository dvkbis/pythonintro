```python
import re

# a = re.findall(r"(\d+)(\[\D+\])", "3[a2[c]]") #"3[a]2[bc]
# a = re.findall(r"(\]|\d+|[a-z]+\]?)","3[a2[c]]" ) # accacacc
# a = re.findall(r"(\]|\d+|[a-z]+\]?)","3[a]2[bc]" )
# a = re.findall(r"(\]|\d+|[a-z]+\]?)","2[abc]3[cd]ef")
#a = re.findall(r"(\]|\d+|[a-z]+\]?)","3[a2[c]]efd4[ab]" ) 
a = re.findall(r"(\]|\d+|[a-z]+\]?)","abc3[2[e]fgh]" ) #abceefgheefgheefgh

result = ""
counters = []
cur_word = ""
for match in a:
    if match.isnumeric():
        counters.append(int(match))
    elif match == ']':
        c = counters.pop()
        cur_word = cur_word * c
    else:
        if match[-1] == ']':
            c = counters.pop()
            cur_word = "".join([cur_word, match[:-1] * c])
        else:
            cur_word = "".join([cur_word, match])
        
    if len(counters) == 0:
        result += cur_word
        cur_word = ""
    
if cur_word:
    result += cur_word

print(result) 
# 3[a2[c]]efd4[ab]
# acc
```
