```python
import re

# a = re.findall(r"(\d+)(\[\D+\])", "3[a2[c]]") #"3[a]2[bc]
a = re.findall(r"(\[|\]|\d+|[a-z]+)","3[a2[c]]" )
#a = re.findall(r"(\[|\]|\d+|[a-z]+)","3[a]2[bc]" )
# a = re.findall(r"(\[|\]|\d+|[a-z]+)","2[abc]3[cd]ef")

result = ""
counters = []
words = []
for b in a:
    if b.isnumeric():
        counters.append(int(b))
    elif b == "[":
        pass
    elif b == "]":
        word = words.pop() * counters.pop()
        if len(counters) > 0:
            words.append(word)
        else:
            result += word
    else:
        words.append(b)
    
if len(words) > 0:
    result += words.pop()

print(result)
# for b in a:
#     print(int(b[0]) * b[1])
```
