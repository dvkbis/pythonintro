```python
class Solution:
    def reverseVowels(self, s: str) -> str:
        left = 0
        right = len(s) - 1
        result = list(s)
        vowels = {'a','e','i','o','u','A','E','I','O','U'}
        while left < right:
            if s[left] in vowels and s[right] in vowels:
                result[left]= s[right]
                result[right] = s[left]
                left += 1
                right -= 1
            else:
                if s[left] not in vowels:
                    left += 1
                else:
                    right -= 1
        
        return "".join(result)
        
```

```python
# Solution Attendue
class Solution:
    def reverseVowels(self, s: str) -> str:
        chars = list(s)
        vowels = set("aeiouAEIOU")

        left = 0
        right = len(chars) - 1

        while left < right:

            while left < right and chars[left] not in vowels:
                left += 1

            while left < right and chars[right] not in vowels:
                right -= 1

            chars[left], chars[right] = chars[right], chars[left]

            left += 1
            right -= 1

        return "".join(chars)
```
