```python
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        a = 0 
        b = len(matrix) - 1 # 2

        while a < b:
            for i in range(a, b - a):
                tmp = matrix[a][a+i]
                matrix[a][a+i] = matrix[b-i][a]
                matrix[b-i][a] = matrix[b][b-i]
                matrix[b][b-i] = matrix[a+i][b]
                matrix[a+i][b] = tmp
            a += 1
            b -= 1

solution = Solution()
solution.rotate([[2,29,20,26,16,28],[12,27,9,25,13,21],[32,33,32,2,28,14],[13,14,32,27,22,26],[33,1,20,7,21,7],[4,24,1,6,32,34]])
```
