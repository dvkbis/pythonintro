```python
from collections import Counter
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        cnt_list = list(Counter (nums))
        for i in range(len(cnt_list)):
            nums[i] = cnt_list[i]
        return len(cnt_list)
```

```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        i = 0
        for val in nums[1:]:
            if nums[i] == val:
                continue
            i += 1
            nums[i] = val


        return i + 1
```
