```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        if len(s) > len(t):
            return False
        i1 = 0
        i2 = 0

        while i2 < len(t) and i1 < len(s):
            if s[i1] == t[i2]:
                i1 += 1
            i2 += 1

        return i1 == len(s)
```

```python
## Plus lisible
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        i = 0

        for c in t:
            if i < len(s) and c == s[i]:
                i += 1

        return i == len(s)
```

```python
# Pythonique
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        it = iter(t)
        return all(c in it for c in s)
```
