```python
def spiralOrder( matrix: List[List[int]]) -> List[int]:
    m = len(matrix)
    n = len(matrix[0])

    result = []

    x = 0
    y = 0
    round = 0
    nb_elements = m * n
    while len(result) < nb_elements:
        min_x, max_x = round, n - round
        min_y, max_y = round, m - round
        y = min_y
        x = min_x
        while y < max_y:
            result.append(matrix[x][y])
            y += 1
        y -= 1
        x += 1
        while x < max_x:
            result.append(matrix[x][y])
            x += 1
        x -= 1
        y -=1
        while y >= min_y:
            result.append(matrix[x][y])
            y -= 1
        y +=1
        x -=1
        while x >= min_x:
            result.append(matrix[x][y])
            x -= 1

        round +=1
```
