```python
from functools import cache
class Solution:
    def canJump(self, nums: List[int]) -> bool:
        n = len(nums)

        @cache
        def jump(i):
            if i == n - 1:
                return True
            if i >= n:
                return False
            nb_jumps = nums[i]
            for j in range(nb_jumps, 0, -1):
                if jump(i + j): 
                    return True

            return False
        
        return jump(0)
            
        
```

```python
# Solution attendue : Greedy O(n)
# nums = [2,3,1,1,4]
class Solution:
    def canJump(self, nums: List[int]) -> bool:
        max_reach = 0

        for i in range(len(nums)):
            if i > max_reach:
                return False

            max_reach = max(max_reach, i + nums[i])

        return True
    
s = Solution()
print(s.canJump([2,3,1,1,4]))
```
