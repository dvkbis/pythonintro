```python
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        if nums[0] > target:
            return 0
        if nums[-1] < target:
            return len(nums)

        start = 0
        end = len(nums) - 1

        while start not in (end, end - 1):
            mid = (start + end ) // 2

            if target <= nums[mid]: 
                end = mid           
            if target >= nums[mid]: 
                start = mid         

        if nums[start] == target:
            return start
        
        return end
        
```
