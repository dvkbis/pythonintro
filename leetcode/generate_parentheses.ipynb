```python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        result = []
        parentheses = { "(" : 1, ")" : 0}
        # count open parenthese
        # close parenthese
        def backtracking(dict_parentheses, combination):
            if len(combination) == n * 2:
                if combination[-1] == ")":
                    result.append("".join(combination))
   
            if dict_parentheses["("] < n:
                combination.append("(")
                dict_parentheses["("] += 1
                backtracking(dict_parentheses, combination)
                combination.pop()
                dict_parentheses["("] -= 1

            if dict_parentheses[")"] < n and dict_parentheses[")"] < dict_parentheses["("]:
                combination.append(")")
                dict_parentheses[")"] += 1
                backtracking(dict_parentheses, combination)
                combination.pop()   
                dict_parentheses[")"] -= 1
            

                        
        backtracking(parentheses, ["("])
        return result
```
