```python
class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        # w1 = abc
        # w2 = ros -> 3x (replace)
        # w1 = abc / w2 = aos -> 2x (replace)
        # w1 = abc / w2 = abo -> 3x (replace ou insert/rm)
        # w1 = abc / w2 = abcd  || abc -> abc[d]
        # w1 = abc / w2 = abdd  || abc -> abc[d] -> abdd
        # w1 = abc / w2 = baddc || abc -> [b]abc -> bab[d]c -> baddc
        # w1 = ros / w2 = horse || ros -> [h]ros[e] -> h[or]se
        # w1 = horse / w2 = ros || horse -> rorse -> rose -> ros
        # w1 = horse / w2 = r   || horse -> orse -> rse -> r
        # w1 = horse / w2 = ds  || horse -> ors -> rs -> ds
        # w1 = horse / w2 = ros || horse -> ors -> ros
        # w1 = abc / w2 = cdabc -> 
        # intention - execution -> entention -> extention -> exeention -> execntion -> execution
        i1 = 0
        i2 = 0
        start2 = 0
        n1 = len(word1)
        n2 = len(word2)
        count = 0
        if n1 == n2:
            for i in range(n1):
                if word1[i] != word2[i]:
                    count += 1 
        else:
            count = max(n1, n2)
            while i1 < n1 :
                i2 = start2
                while i2 < n2:
                    if word1[i1] == word2[i2]:
                        count -= 1
                        start2 = i2 + 1
                        i2 = n2
                    else:
                        i2 += 1
                i1 += 1
        

        return count


        
```
