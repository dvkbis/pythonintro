```python
import string

class Solution:
    def decodeMessage(self, key: str, message: str) -> str:
        key_dict = {" ": " "}
        alphabet = string.ascii_lowercase
        result = []
        i_key = 0
        for c in alphabet:
            while i_key < len(key) and key[i_key] in key_dict:
                i_key += 1


            key_dict[ key[i_key]] = c

        for c in message:
            result.append(key_dict[c])

        return "".join(result)

            



        
        
```
