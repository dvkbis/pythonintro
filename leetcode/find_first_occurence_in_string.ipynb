```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if len(needle) > len(haystack):
            return -1
        # input = needle haystack
        # -> i of needle in haystack
        # -> -1 if no needle in haystack
        i1 = 0
        i2 = 0
        found = False

        while (len(haystack) - i1) >= len(needle) and not found:
            i2 = 0
            while i2 < len(needle) and haystack[i1] == needle[i2]:
                i1 += 1
                i2 += 1

            if i2 == len(needle):
                found = True
            else:
                i1 = i1 - i2 + 1

        return i1 - i2 if i2 == len(needle) else -1
```

```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if len(needle) > len(haystack):
            return -1

        i1 = 0

        while (len(haystack) - i1) >= len(needle):
            i2 = 0
            prev = i1

            while i2 < len(needle) and haystack[i1] == needle[i2]:
                i1 += 1
                i2 += 1

            if i2 == len(needle):
                return prev
            else:
                i1 = prev + 1

        return -1


```

```python
    
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if len(needle) > len(haystack):
            return -1

        start = 0

        while (len(haystack) - start) >= len(needle):
            end = start + len(needle)
            if haystack[start:end] == needle:
                return start
            start += 1

        return -1

solution = Solution()
# "sadbutsad", needle = "sad"
solution.strStr("hellosadbutsad", "sad")
```
