```python
def maxSubArray(nums: List[int]) -> int:
    if len(nums) == 0:
        return []
    ranges, max_sub = max_sub_array_helper(nums, 0, len(nums) - 1) 
    return nums[ranges[0]: ranges[1] + 1]

def max_sub_array_helper(nums, left, right):
    # Base case
    if left == right:
        return (left, right), nums[left]
    
    # DIVIDE
    mid = (left + right) // 2

    # CONQUER
    left_max = max_sub_array_helper(nums, left, mid) 
    right_max = max_sub_array_helper(nums, mid + 1, right)

    # COMBINE
    mid_max = crossing_max(nums, left, mid, right)

    best_max = compare_max_sub(left_max, right_max)  
    return compare_max_sub(best_max, mid_max)
def crossing_max(nums, left, mid, right):
    max_left, index_left = find_max(nums, mid, left - 1, -1)
    max_right, index_right = find_max(nums, mid+1, right + 1, 1)
    sum_left_max = max_left + max_right
    return (index_left, index_right), sum_left_max
        
def find_max(nums, start, end, step):
    max_value = nums[start]
    end_index = start
    current_sum = max_value

    for i in range(start + step, end, step):
        current_sum += nums[i]
        if current_sum > max_value:
            max_value = current_sum
            end_index = i
    return max_value, end_index

def compare_max_sub(left, right):

    if left[1] > right[1]:
        return left
    elif left[1] == right[1]:
        left_size = left[0][1] - left[0][0]
        right_size = right[0][1] - right[0][0]
        if left_size >= right_size:
            return left
        else:
            return right
    else:
        return right


print(maxSubArray( [-2,1,-3,4,-1,2,1,-5,4])) #  [4,-1,2,1]
print(maxSubArray( [5,4,-1,7,8] )) # [5,4,-1,7,8]
print(maxSubArray( [1]))
```

```python
class Solution:
    # EX : [-2,1,-3,4,-1,2,1,-5,4]
    # [-2]
    # [1] *
    # [1, -3]
    # [4] *
    # [4, -1]
    # [4, -1, 2] = 5 *
    # [4, -1, 2, 1] = 6 *
    # [4, -1, 2, 1, -5] = 1
    # [4, -1, 2, 1, -5, 4] = 5
    # 
    # Si un nombre isolé dépasse le total : on commence part ce nombre 
    def maxSubArray(self, nums: List[int]) -> int:


        largest_sum = nums[0]
        cur_sum = nums[0]
        for n in nums[1:]:
            cur_sum = max(n, cur_sum + n)
            largest_sum = max(largest_sum, cur_sum)

        return largest_sum
        
                
```
