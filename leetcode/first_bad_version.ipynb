```python
# The isBadVersion API is already defined for you.
# def isBadVersion(version: int) -> bool:

class Solution:
    # 10
    def firstBadVersion(self, n: int) -> int:
        if n == 2:
            return 1 if isBadVersion(1) else 2

        start = 1   # 1
        end = n     # 2
        while end not in(start, start + 1):
            mid = (start + end) // 2 # 8

            if isBadVersion(mid):
                end = mid 
            else:
                start = mid

        return start if isBadVersion(start) else end
```
