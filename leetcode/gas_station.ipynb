```python
class Solution:
    def canCompleteCircuit(self, gas: List[int], cost: List[int]) -> int:
        # Input: gas = [1,2,3,4,5], cost = [3,4,5,1,2]
        # i | cost | gain 
        # 0 |  -3  |  1     -2
        # 1 |  -4  |  2     -2
        # 2 |  -5  |  3     -2
        # 3 |  -1  |  4      3
        # 4 |  -2  |  5      3

    # 5 > -2 > -2 > -2 > 3 > 3 > -
    # 5 > -1 >  2 > 
    # 3
    # 5 > -3 > 3 > - 4

        # Input: gas = [2,3,4], cost = [3,4,3]
        # i | cost | gain 
        # 0 |  -3  |  2      -1
        # 1 |  -4  |  3      -1
        # 2 |  -3  |  4      -1

        # Input: gas = [0,0,10,1], cost = [10,1,1,1]
        # i | cost  | gain 
        # 0 |  -10  |  0      -10
        # 1 |    1  |  0      -1
        # 2 |    1  |  10     9
        # 3 |    1  |  1      0

        # Trouver la suite positive la plus grande?
        # Total de tous les voyages : si < 0 : return -1
        # best_total_positif = 0
        total = 0
        best = -1
        for i in range(len(gas)):
            usage = gas[i] - cost[i]
            total += usage
            if total < 0:
                best = -1
            elif best == -1:
                best = max(usage, best)

        print(best, total)
        if total >= 0:
            return best - 1 if best != 0 else len(gas) -1
        return -1


        # Input: gas = [5,1,2,3,4], cost = [4,4,1,5,1]
        # i | cost | gain 
        # 0 |  -4  |  5       1
        # 1 |  -4  |  1      -3
        # 2 |  -1  |  2       1
        # 3 |  -5  |  3      -2
        # 4 |  -1  |  4       3
```
