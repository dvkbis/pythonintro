```python
def isValidSudoku(board: List[List[str]]) -> bool:
    my_dict = {}
# 5 = (0,0), (7,8), (1,5)
    for i in range(9):
        for j in range(9):
            value = board[i][j]
            if value == ".":
                continue
            if value in my_dict:
                my_dict[value].append((i,j))
            else:
                my_dict[value] = [(i,j)]
    print (my_dict)
    for key, coords in my_dict.items():
        print("----", key)
        x_list = set()
        y_list = set()
        sub_boxes = set()

        for x, y in coords:
            if x in x_list:
                print("invalid x")
                return False
            if y in y_list:
                print("invalid y")
                return False
            sub_box = (x // 3, y // 3)
            if sub_box in sub_boxes:
                print("invalid sub_box")
                return False
            x_list.add(x)
            y_list.add(y)
            sub_boxes.add(sub_box)

    return True

board = [["5","3",".",".","7",".",".",".","."],["6",".",".","1","9","5",".",".","."],[".","9","8",".",".",".",".","6","."],["8",".",".",".","6",".",".",".","3"],["4",".",".","8",".","3",".",".","1"],["7",".",".",".","2",".",".",".","6"],[".","6",".",".",".",".","2","8","."],[".",".",".","4","1","9",".",".","5"],[".",".",".",".","8",".",".","7","9"]]
print(isValidSudoku(board))
```
