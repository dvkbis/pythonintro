```python
class Solution:
    def getRow(self, rowIndex: int) -> List[int]:
        prev = [1]
        for i in range(0, rowIndex):
            row = []
            row.append(1)
            for j in range(1, len(prev)):
                row.append(prev[j] + prev[j-1])
            row.append(1)
            prev = row
        
        return prev
```
