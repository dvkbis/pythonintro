```python
class Solution:
    def partitionLabels(self, s: str) -> List[int]:
        char_map = {}
        # Init index range of each characters
        for i in range(len(s)):
            c = s[i]
            if c not in char_map:
                char_map[c] = [i,i]
            else:
                char_map[c][1] = i

        part = [-1, -1]
        result = []
        for v in char_map.values():
            # check if a char is inside the current partition
            if part[0] < v[0] < part[1]:
                if v[1] > part[1]:
                    part[1] = v[1]
            # add the partition
            else:
                result.append(part[1] - part[0] + 1)
                part.clear()
                part = v
        # add last partition
        if part:
            result.append(part[1] - part[0] + 1)

        result.pop(0)
        return result
            
```
