```python
def solveSudoku(board: List[List[str]]) -> None:
    """
    Do not return anything, modify board in-place instead.
    """
    # fill sudo with possible numbers
    # list all completed number in 
    #   - col
    #   - row
    #   - bloc
    hor = []
    ver = []
    bloc_list = []
    # Init
    for i in range(9):
        hor.append(set())
        ver.append(set())
    for i in range(3):
        bloc = []
        for j in range(3):
            bloc.append(set())
        bloc_list.append(bloc)

    # fill
    for i in range(9):
        for j in range(9):
            if board[i][j] == ".":
                continue
            hor[i].add(board[i][j])
            ver[j].add(board[i][j])
            bloc_i = i // 3
            bloc_j = j // 3
            bloc_list[bloc_i][bloc_j].add(board[i][j])
    
    
    print("hor", hor)
    print("ver", ver)
    print("bloc", bloc_list)
    possible={1,2,3,4,5,6,7,8,9}
    draft = []
    for i in range(9):
        row = []
        draft.append(row)
        for j in range(9):
            if board[i][j] != ".":
                draft_cell = hor[i] | ver[j] | bloc_list[i // 3][j // 3]
                left = possible.symmetric_difference(draft_cell)
                if len(left) == 1:
                    board[i][j] = left[0]
                # fill ver / hor /bloc
                row.append(left)
            else:
                row.append(set(board[i][j]))

    
    print(draft)

    # draft -> resolution -> draft -> resolution
    
board = [["5","3",".",".","7",".",".",".","."],["6",".",".","1","9","5",".",".","."],[".","9","8",".",".",".",".","6","."],["8",".",".",".","6",".",".",".","3"],["4",".",".","8",".","3",".",".","1"],["7",".",".",".","2",".",".",".","6"],[".","6",".",".",".",".","2","8","."],[".",".",".","4","1","9",".",".","5"],[".",".",".",".","8",".",".","7","9"]]
output = [["5","3","4","6","7","8","9","1","2"],["6","7","2","1","9","5","3","4","8"],["1","9","8","3","4","2","5","6","7"],["8","5","9","7","6","1","4","2","3"],["4","2","6","8","5","3","7","9","1"],["7","1","3","9","2","4","8","5","6"],["9","6","1","5","3","7","2","8","4"],["2","8","7","4","1","9","6","3","5"],["3","4","5","2","8","6","1","7","9"]]
solve = solveSudoku(board)
print(solve)  
# print(board == solve)

a = {1,2,3,4}
b = {2,3,4}
print(a.symmetric_difference(b))
```
