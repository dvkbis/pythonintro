```python
def solveSudoku(board: List[List[str]]) -> None:
    """
    Do not return anything, modify board in-place instead.
    """
    # fill sudo with possible numbers
    # list all completed number in 
    #   - col
    #   - row
    #   - bloc
    hor = []
    ver = []
    bloc_list = []
    # Init
    for i in range(9):
        hor.append(set())
        ver.append(set())
    for i in range(3):
        bloc = []
        for j in range(3):
            bloc.append(set())
        bloc_list.append(bloc)

    # fill
    for i in range(9):
        for j in range(9):
            if board[i][j] == ".":
                continue
            hor[i].add(board[i][j])
            ver[j].add(board[i][j])
            bloc_i = i // 3
            bloc_j = j // 3
            bloc_list[bloc_i][bloc_j].add(board[i][j])
    
    
    print("hor", hor)
    print("ver", ver)
    print("bloc", bloc_list)
    possible={"1","2","3","4","5","6","7","8","9"}
    draft = []
    changed = True
    t = 0
    while not is_resolve(board):
        if not changed: 
            break
        # changed = False
        new_draft = []
        for i in range(9):
            row_draft = []
            new_draft.append(row_draft)
            for j in range(9):
                if board[i][j] == ".":
                    draft_cell = hor[i] | ver[j] | bloc_list[i // 3][j // 3]
                    remaining = possible.symmetric_difference(draft_cell)
                    if len(remaining) == 1:
                        board[i][j] = next(iter(remaining))
                        hor[i].add(board[i][j])
                        ver[j].add(board[i][j])
                        bloc_list[i // 3][j // 3].add( board[i][j])
                        
                    # fill ver / hor /bloc
                    row_draft.append(remaining)
                    hor[i] = remaining
                    ver[j] = remaining
                    bloc_list[i // 3][j // 3] = remaining
                else:
                    row_draft.append(set(board[i][j]))

        draft = new_draft
        if t >= 5:
            for line in board:
                print(" ".join(line))         
            break
        t+=1
                # if len(draft[i][j]) == 0:
                #     print("here")
                #     return
        print("draft", draft)
        print("board", board)

def is_resolve(board):
    resolved = True
    i = 0
    while resolved and i < 9:
        j = 0
        while resolved and j < 9:
            if board[i][j] == ".":
                resolved = False
            j += 1
        i +=1  
    return resolved

 
board = [["5","3",".",".","7",".",".",".","."],["6",".",".","1","9","5",".",".","."],[".","9","8",".",".",".",".","6","."],["8",".",".",".","6",".",".",".","3"],["4",".",".","8",".","3",".",".","1"],["7",".",".",".","2",".",".",".","6"],[".","6",".",".",".",".","2","8","."],[".",".",".","4","1","9",".",".","5"],[".",".",".",".","8",".",".","7","9"]]
output = [["5","3","4","6","7","8","9","1","2"],["6","7","2","1","9","5","3","4","8"],["1","9","8","3","4","2","5","6","7"],["8","5","9","7","6","1","4","2","3"],["4","2","6","8","5","3","7","9","1"],["7","1","3","9","2","4","8","5","6"],["9","6","1","5","3","7","2","8","4"],["2","8","7","4","1","9","6","3","5"],["3","4","5","2","8","6","1","7","9"]]
print(board)
# solve = solveSudoku(board)
# print(solve)  
# print(board)
# print(output)
# print(board == output)

board = [[".",".","9","7","4","8",".",".","."],["7",".",".",".",".",".",".",".","."],[".","2",".","1",".","9",".",".","."],[".",".","7",".",".",".","2","4","."],[".","6","4",".","1",".","5","9","."],[".","9","8",".",".",".","3",".","."],[".",".",".","8",".","3",".","2","."],[".",".",".",".",".",".",".",".","6"],[".",".",".","2","7","5","9",".","."]]
print(board)
solveSudoku(board)
print(board)
for line in board:
    print(" ".join(line))
# SI DOUBLE/ TRIPLE/... sur HOR / VER / BLOC
# -> [0,0]= {9}        [0,1]={9}        [0,2]={/}
# -> [1,0]= {9}        [1,1]={9}        [1,2]={9}
# -> [2,0]= {/}        [2,1]={9}        [2,2]={9}

# . . 9 | 7 4 8 |. . .
# 7 . . | 6 . 2 |. . .
# . 2 . | 1 . 9 |. . .
#----------------------
# . . 7 | 9 8 6 |2 4 1
# 2 6 4 | 3 1 7 |5 9 8
# 1 9 8 | 5 2 4 |3 6 7
#----------------------
# . . . | 8 6 3 |. 2 .
# . . . | 4 9 1 |. . 6
# . . . | 2 7 5 |9 . .
```
