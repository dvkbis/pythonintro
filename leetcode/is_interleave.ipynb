```python
from functools import cache

def isInterleave(s1: str, s2: str, s3: str) -> bool:
    if len(s1) + len(s2) != len(s3):
        return False
    
    @cache
    def dfs(i1, i2, i3):
        if i3 == len(s3):
            return True
        
        is_inter_leave = False
        if i1 < len(s1) and s1[i1] == s3[i3]:
            is_inter_leave = dfs(i1 + 1, i2, i3 + 1)

        if not is_inter_leave: 
            if i2 < len(s2) and s2[i2] == s3[i3]:
                is_inter_leave = dfs(i1, i2 + 1, i3 + 1)
        
        return is_inter_leave

    return dfs(0, 0, 0)


print("False", isInterleave("aabcc", "dbbca", "aadbbbaccc")) 
print("True", isInterleave("aabcc", "dbbca", "aadbbcbcac")) 
print(isInterleave("bbbbbabbbbabaababaaaabbababbaaabbabbaaabaaaaababbbababbbbbabbbbababbabaabababbbaabababababbbaaababaa",
                           "babaaaabbababbbabbbbaabaabbaabbbbaabaaabaababaaaabaaabbaaabaaaabaabaabbbbbbbbbbbabaaabbababbabbabaab",
                             "babbbabbbaaabbababbbbababaabbabaabaaabbbbabbbaaabbbaaaaabbbbaabbaaabababbaaaaaabababbababaababbababbbababbbbaaaabaabbabbaaaaabbabbaaaabbbaabaaabaababaababbaaabbbbbabbbbaabbabaabbbbabaaabbababbabbabbab")) 
```
