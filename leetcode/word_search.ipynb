```python
def exist(board: List[List[str]], word: str) -> bool:
    m = len(board)
    n = len(board[0])

    visited = []
    def is_good_letter(index_word, i, j ):
        if index_word == len(word):
            return True
        if i < 0 or i >= m:
            return False
        if j < 0 or j >= n:
            return False
        if word[index_word] != board[i][j]:
            return False
        if (i, j) in visited:
            return False
        
        visited.append((i,j))

        if (is_good_letter(index_word + 1, i + 1, j)):
            return True
        if (is_good_letter(index_word + 1, i, j + 1)):
            return True
        if (is_good_letter(index_word + 1, i - 1, j)):
            return True
        if (is_good_letter(index_word + 1, i, j - 1)):
            return True

        visited.pop()
    

    for i in range(m):
        for j in range(n):     
            if (is_good_letter(0, i, j)):
                return True
    return False



        
```
