```python
def exist(board: List[List[str]], word: str) -> bool:
    m = len(board)
    n = len(board[0])

    visited = []
    def is_good_letter(index_word, i, j ):
        if index_word == len(word):
            return True
        if i < 0 or i >= m:
            return False
        if j < 0 or j >= n:
            return False
        if word[index_word] != board[i][j]:
            return False
        if (i, j) in visited:
            return False
        
        visited.append((i,j))

        if (is_good_letter(index_word + 1, i + 1, j)):
            return True
        if (is_good_letter(index_word + 1, i, j + 1)):
            return True
        if (is_good_letter(index_word + 1, i - 1, j)):
            return True
        if (is_good_letter(index_word + 1, i, j - 1)):
            return True

        visited.pop()
    

    for i in range(m):
        for j in range(n):     
            if (is_good_letter(0, i, j)):
                return True
    return False
```

```python
# Solution Attendue
class Solution:
    def exist(self, board, word):
        rows = len(board)
        cols = len(board[0])

        visited = set()

        def dfs(r, c, k):

            if k == len(word):
                return True

            if (
                r < 0
                or r >= rows
                or c < 0
                or c >= cols
                or board[r][c] != word[k]
                or (r, c) in visited
            ):
                return False

            visited.add((r, c))

            for dr, dc in [(1,0),(-1,0),(0,1),(0,-1)]:
                if dfs(r + dr, c + dc, k + 1):
                    return True

            visited.remove((r, c))
            return False

        for r in range(rows):
            for c in range(cols):
                if dfs(r, c, 0):
                    return True

        return False
```
