```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        buy = prices[0]
        sell = 0
        for n in prices[1:]:
            buy = min (buy, n) 
            sell = max (sell, n - buy) 
        return sell
        
```

```python
# Solution attendue : Greedy
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        min_price = prices[0]
        max_profit = 0

        for price in prices[1:]:
            min_price = min(min_price, price)
            max_profit = max(max_profit, price - min_price)

        return max_profit
```
