```python
def lengthOfLongestSubstring(s: str) -> int:
    n = len(s)
    if n == 0:
        return 0
    letter_index = {}
    max_substring = -1
    count = 0
    for i in range(n):
        char = s[i]
        if char in letter_index:
            max_substring = max(max_substring, count)
            count = i - letter_index[char] - 1
            
        count += 1
        letter_index[char] = i
        print(count, max_substring)
    
    if count > max_substring:
        return count
    else:
        return max_substring

print(lengthOfLongestSubstring("abba"))
# 
```

```python
def lengthOfLongestSubstring(s: str) -> int:
    cur = ""
    best = 0
    for c in s:
        if c in cur:

            cur = cur.split(c,1)[1]

        cur += c
        best = max(len(cur), best)

    return best
print(lengthOfLongestSubstring("abba"))
print(lengthOfLongestSubstring("abacdea"))
print(lengthOfLongestSubstring("aaaaaa"))

```
